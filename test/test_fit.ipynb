```python
from pathlib import Path
from typing import Sequence

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from pyama_core.utils.csv_loader import load_csv_data
from pyama_core.analysis.utils.trace_fitting import fit_trace_data
from pyama_core.analysis.models import get_model
```

```python
df = load_csv_data('../data/HuH7_FCSnew_GFP.csv')
```

```python
cell_id = np.random.randint(0, df.shape[1])
res = fit_trace_data(df, model_type='trivial', cell_id=cell_id)
t = df.index.values.astype(float)
y = df.iloc[:, cell_id].values.astype(float)
model_mod = get_model('trivial')
params = res.fitted_params
y_pred = model_mod.eval(t, params)
plt.title(f'cell {cell_id}')
plt.plot(t, y, 'o', ms=3)
plt.plot(t, y_pred, '-', lw=2)
plt.show()
```
